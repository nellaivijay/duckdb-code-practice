## 🚀 Step 2: Connection Testing Framework

This notebook provides a framework for testing MotherDuck connections once you have an account.

```python
import duckdb
import os

# Framework for MotherDuck connection
# Uncomment and fill in your actual token when ready

# os.environ['MOTHERDUCK_TOKEN'] = 'your_token_here'

def test_motherduck_connection():
    """Test MotherDuck connection"""
    try:
        con = duckdb.connect('md:')
        result = con.execute("SELECT current_database()").fetchone()
        print(f"✓ Connected to MotherDuck: {result[0]}")
        
        # List tables
        tables = con.execute("SHOW TABLES").fetchall()
        print(f"✓ Tables in database: {len(tables)}")
        
        con.close()
        return True
    except Exception as e:
        print(f"✗ Connection failed: {e}")
        return False

# Test connection (will fail without actual token)
print("Testing MotherDuck connection framework...")
test_motherduck_connection()
```

## 🚀 Step 3: Local Database Simulation

Simulate MotherDuck operations using local DuckDB for learning purposes.

```python
# Simulate MotherDuck operations locally
con = duckdb.connect('data/motherduck_simulation.db')

# Create sample data
con.execute("""
    CREATE TABLE cloud_sales AS
    SELECT 
        (random() * 1000)::int as customer_id,
        (random() * 100)::int as product_id,
        random() * 500 as amount,
        '2023-01-01'::date + (random() * 365)::int as sale_date
    FROM range(10000)
""")

print("✓ Created simulated cloud database with 10,000 sales records")

# Simulate cloud queries
result = con.execute("""
    SELECT 
        customer_id,
        COUNT(*) as order_count,
        SUM(amount) as total_amount
    FROM cloud_sales
    GROUP BY customer_id
    ORDER BY total_amount DESC
    LIMIT 10
""").fetchdf()

print("\nTop 10 customers by total amount:")
print(result)
```

## 🚀 Step 4: Data Upload Simulation

Simulate uploading data to MotherDuck.

```python
# Simulate data upload process
local_con = duckdb.connect('data/duckdb_practice.db')

# Export data for "upload"
local_con.execute("""
    COPY sample_customers TO 'data/customers_upload.parquet' (FORMAT PARQUET)
""")

# Simulate cloud database creation and data loading
cloud_con = duckdb.connect('data/motherduck_simulation.db')

cloud_con.execute("""
    CREATE TABLE uploaded_customers AS
    SELECT * FROM 'data/customers_upload.parquet'
""")

count = cloud_con.execute("SELECT COUNT(*) FROM uploaded_customers").fetchone()[0]
print(f"✓ Simulated upload: {count} customers loaded to cloud database")
```

## 🚀 Step 5: S3 Integration Framework

Framework for S3 integration (requires actual S3 credentials).

```python
def setup_s3_integration(con, access_key, secret_key, region='us-east-1'):
    """Set up S3 integration for DuckDB"""
    try:
        con.execute(f"""
            CREATE SECRET s3_secret (
                TYPE S3,
                KEY_ID '{access_key}',
                SECRET '{secret_key}',
                REGION '{region}'
            )
        """)
        print("✓ S3 secret configured")
        return True
    except Exception as e:
        print(f"✗ S3 setup failed: {e}")
        return False

# Framework function (requires actual credentials)
print("S3 integration framework ready")
print("To use: setup_s3_integration(con, 'your_key', 'your_secret', 'us-east-1')")
```
